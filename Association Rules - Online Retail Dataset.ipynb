### 5. Model Building

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

### Get Best Assocaites for France

```python
france_data =retail_data[retail_data['Country']== 'France']
france_data
```

```python
france_Data_pivot =pd.pivot_table(data = france_data,values= 'Quantity',index= 'InvoiceNo', columns= 'Description').fillna(0)
france_Data_pivot
```

```python
france_encoded =france_Data_pivot.applymap(lambda  x:1 if x>0 else 0)
france_encoded
```

```python
france_freq_itemsets =apriori(df= france_encoded,min_support= 0.02,use_colnames=True)
france_freq_itemsets
```

```python
france_best_associate = association_rules(df = france_freq_itemsets, metric='confidence',min_threshold= 0.7)
france_best_associate
```

### Task get the associates for Germany

```python
germany_data =retail_data[retail_data['Country']== 'Germany']
germany_data
```

```python
germany_pivot=pd.pivot_table(data= germany_data,values='Quantity', index='InvoiceNo',columns='Description').fillna(0)
germany_pivot
```

```python
germany_encoded= germany_pivot.applymap(lambda x:1 if x>1 else 0)
germany_encoded
```

```python
germany_freq_itemsets = apriori(df= germany_encoded,min_support=0.01,use_colnames=True,)
germany_freq_itemsets
```

```python
germany_best_Associates = association_rules(df= germany_freq_itemsets,metric='confidence',min_threshold=0.9,)
germany_best_Associates
```

```python
germany_best_Associates.sort_values(by=(['confidence','antecedents']),ascending=False,)
```

```python
germany_50_best_associate =germany_best_Associates.sort_values(by=(['confidence','antecedents']),ascending=False,).head(50)
germany_50_best_associate
```
